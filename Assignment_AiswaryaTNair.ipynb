# 1.check whether there is any relationship between the amount spent for advertisement through five different companies and profit generated.

```python
corrmatrix=sales_data.corr()
plt.subplots(figsize=(6,6))
sns.heatmap(corrmatrix,vmin=-.7,vmax=1,linewidth=.20,annot=True)
```

```python
# adding column total amount spent for advertisement.
sales_data['total amount spent for advertisement']=sales_data.iloc[:,0:5].sum(axis=1)
```

```python
sales_data.head()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(sales_data['total amount spent for advertisement'],sales_data['Profit(in Thousands)'],s=15,c='red',marker='^')
plt.title('total amount spent for advertisement vs profit',fontsize=15)
plt.xlabel('total amount spent for advertisement')
plt.ylabel('Profit(in Thousands)')
```

# insight
1.This graph shows that there is a strong positive correlation between profit and total amount spent for advertisement.

# 2. Build a regression model to predict the profit from the given data


```python

data1=sales_data.drop(['Add1(in Thousands)','Add5(in Thousands)','total amount spent for advertisement'],axis=1)
```

```python
data1
```

```python
#dependant variable is profit(y) and others are the independant variable
y=data1['Profit(in Thousands)']
x=data1.drop(['Profit(in Thousands)'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.9)
```

```python
from sklearn import linear_model
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=linear_model.LinearRegression()
```

```python

model=lr.fit(x_train,y_train)
      
```

```python
predictions=model.predict(x_test)
```

```python
print(y_test)
```

```python
print(predictions)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print('MSE is',mean_squared_error(y_test,predictions))
```
